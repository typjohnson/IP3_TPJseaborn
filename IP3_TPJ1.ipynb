```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).astype('int64')
```

```python
car_sales.info()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
num_rows = len(car_sales)
today = pd.Timestamp.now().date()
date_range = pd.date_range(start=today, periods=num_rows)
car_sales['Sale Date'] = date_range
```

```python
plt.figure(figsize=(12, 6))
plt.plot(car_sales['Price'], car_sales['Total Sales'], label='Price')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Total Sales vs. Price')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Scatter Plot: Odometer (KM) vs. Price')
plt.show()

sns.scatterplot(x = 'Odometer (KM)',
               y = 'Price',
               data = car_sales,
               ax = axes[1,0])
```
